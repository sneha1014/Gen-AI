```python
from pdf2image import convert_from_path
from paddleocr import PaddleOCR
import numpy as np

# Create an OCR object with angle classification and English language support
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # use_angle_cls=True for better angle correction

def extract_text_from_pdf(pdf_path):
    # Convert PDF to images (one per page)
    images = convert_from_path(pdf_path)

    all_text = ""
    for i, image in enumerate(images):
        # Convert PIL.Image to NumPy array
        image_np = np.array(image)
        
        # OCR processing on the image
        result = ocr.ocr(image_np, cls=True)
        
        # Extract the text from OCR result
        page_text = ""
        for line in result:
            for item in line:
                page_text += item[1][0] + "\n"  # Access the text content from OCR results
        
        all_text += f"--- Page {i + 1} ---\n{page_text}\n"
    
    return all_text

# Example usage
pdf_path = 'C:/Users/SNEHA/OneDrive/Desktop/extraction/Sneha (Resume).pdf'  # Path to the PDF you want to extract text from
extracted_text = extract_text_from_pdf(pdf_path)

# Print the extracted text
#print(extracted_text)


# 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
```

```python
import requests
#from extract import extracted_text

# Set your Groq API endpoint and API key
GROQ_API_URL = "https://api.groq.com/openai/v1/chat/completions"

GROQ_API_KEY = "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"

def map_text_to_keys(text, prompt_instructions):
    """
    Send text to the Groq API and use the Mistral-8x7B model to extract structured information.
    
    :param text: Input text to analyze.
    :param prompt_instructions: Prompt instructions for the model.
    :return: Parsed output as a dictionary.
    """
    headers = {
        "Authorization": f"Bearer {GROQ_API_KEY}",
        "Content-Type": "application/json",
    }
    
    # Construct payload
    payload = {
        "model": "mixtral-8x7b-32768",
        "messages": [
            {"role": "system", "content": prompt_instructions},
            {"role": "user", "content": text}
        ],
        #"prompt": f"{prompt_instructions}\n\n{text}\n\nExtracted Information:",
        "temperature": 0,  # Set low temperature for deterministic responses
        "max_tokens": 512,
    }
    
    # Make the API request
    response = requests.post(GROQ_API_URL, json=payload, headers=headers)
    
    if response.status_code == 200:
        response_data = response.json()
        return response_data["choices"][0]["message"]["content"].strip()
    else:
        raise Exception(f"API request failed with status code {response.status_code}: {response.text}")

prompt_instructions="Extract the applicant name"
try:
    extracted_info = map_text_to_keys(extracted_text, prompt_instructions)
    print(f"Extracted Applicant Name: {extracted_info}")
except Exception as e:
    print(f"Error: {e}")

```
